```python
from bs4 import BeautifulSoup
import requests
import urllib.parse
import pandas as pd
import numpy as np
```

```python
all_web_links=['https://www.trulia.com/CA/San_Diego/','https://www.trulia.com/NM/Albuquerque/']
```

```python
website='https://www.trulia.com/CA/San_Diego/'
```

```python
request=requests.get(website)
```

```python
soup=BeautifulSoup(request.content,'html.parser')
```

```python
allresult_links=soup.find_all('li',{'class':'SearchResultsList__WideCell-b7y9ki-2'})
```

```python
len(allresult_links)
```

```python
data=[]
for i in allresult_links:
    if i.has_attr('data-testid'):
        data.append(i)
```

```python
len(data)
```

```python
all_links=[]
for d in data:
    for link in d.find_all('div',{'data-testid':'property-card-details'}):
        all_links.append('https://www.trulia.com'+link.find('a').get('href'))
```
